1. Indicar si el numero dado es par o impar

```
x = float(input("x = "))
if (-1)**x == 1:
  print ("Par")
else:
  print ("impar")
```

9. Factorial de un numero

```
 x = int(input("Ingrese un numero para calcular el factorial"))
if x == 0:
  print("No es posible esta operacion")
else:
  fact=1
  for i in range (1,x+1):
      fact=fact*i
      print("El factorial del numero dado es:", fact)
```
